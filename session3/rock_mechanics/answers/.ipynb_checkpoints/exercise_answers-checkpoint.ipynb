# Calculating elastic moduli

In rock mechanics, there are multiple ways in which we derive the Young's modulus (elasticity) of a rock sample:
1) Tangential Young's modulus: σ/ε measured at a tangent to the curve at a fixed percentage of the peak stress, normally at 50%

2) Secant Young's modulus: σ/ε measured from zero stress to a fixed percentage of the peak stress, normally 50%

3) Average Young's modulus: average slope of the more-or-less straight line portion of the curve



### Tangential Young's modulus

To calculate tangential Young's modulus, we first want to find the strain value at 50% of the peak stress.

It is logical to think we could use np.where(df.stress=peak/2) to get the index of the value equivalent to 50% peak stress, however in this case we can't do that because the value might not be EXACTLY equal to this number, so we need to find the index of the value closest to this number. To do this we subtract the number we're looking for from every element and find the minimum value.
 
You can test this by looking for the index of the peak stress using both methods.

```python
# establish index of 50% peak stress
idx=(np.abs(df.stress - peak/2)).argmin()
print(idx)
```

```python
# you can double check this result by listing the value and comparing it to the peak stress
df.stress[idx]
```

Because the data is closely spaced, we want to take a datapoint either side of this 50% value to use as the tangent. Effectively what we are doing here is: (change in x)/(change in y), where we are taking the change just above and just below 50% peak stress.

```python
tang_youngs = (df.stress[idx+1]-df.stress[idx-1])/(df.strain[idx+1]-df.strain[idx-1])
print(tang_youngs)
```

```python
print ("The tangential Young's modulus is %.f MPa" %tang_youngs)
```

Typically, we express Young's modulus in GPa. Change the print command to do this.

```python
print ("The tangential Young's modulus is %.2f GPa" %(tang_youngs/1000))
```

### Secant Young's Modulus

 The Secant young's modulus is just σ/ε measured from zero stress to a fixed percentage of the peak stress, normally 50%. Effectively, we are doing this calculation: (change in stress from 0 to 50% peak stress)/(change in strain from 0 to 50% peak stress).

```python
# look up strain value at 50% peak stress
strain_at_50=df.strain[idx]
strain_at_50
```

```python
secant_youngs = (df.stress[idx]-0)/(strain_at_50-0)
print(secant_youngs)
```

Just like before, we can print this to screen in a user-friendly format.

```python
print ("The secant Young's modulus is %.2f GPa" %(secant_youngs/1000))
```

### Average Young's Modulus

To calculate the average Young's modulus, we need to quantitatively establish where the linear portion of the curve is. To do this, we calculate the gradient between each consecutive point for the linear portion, and then take the average of these gradients.

1. Estimate by eye from the plot the start and end x bounds of the linear portion.

2. Write a for loop which cycles through these points, calculating the gradient for each interval.

3. Calculate the mean gradient for this portion of the curve.

```python
# first we need to define the x bounds of the linear portion of the curve, and find the corresponding indices

xmin = 0.002
xmax = 0.003

idxmin=np.abs(df.strain-xmin).argmin()
idxmax=np.abs(df.strain-xmax).argmin()
```

```python
# next we construct the for loop

mod=np.zeros(shape=(idxmax-idxmin-1)) # initialise array mod with length of amount of indices between xmin and xmax minus 1
j=0  # initialise j as 0 - this will control writing values to the array mod

for i in range (idxmin,idxmax-1):
    mod[j]=(df.stress[i+1]-df.stress[i])/(df.strain[i+1]-df.strain[i]) # gradient = change in x/change in y
    j=j+1
    
print(mod)
```

We now have an array that shows us the gradient for every interval between our defined bounds of the linear portion of the curve. We want to calculate the average and the error (1 x standard deviation).

```python
av_youngs=np.mean(mod)
print(av_youngs)
```

```python
av_youngs_std=np.std(mod)
print(av_youngs_std)
```

# Displaying calculated Young's moduli on the plot

First, we want to add a textbox to our plot that tells us the values for each of the Young's modulus calculations. 

Annotate is most commonly used to text at a position (xytext) relative to a point (xy) and optionally connect the text and the point with an arrow, which we'll skip here.

This is a bit complex, so let's break it down:

ax.annotate('...Additional information...',  # Your string

            # The point that we'll place the text in relation to 
            xy=(0.5, 0), 
            # Interpret the x as axes coords, and the y as figure coords
            xycoords=('axes fraction', 'figure fraction'),

            # The distance from the point that the text will be at
            xytext=(0, 10),  
            # Interpret `xytext` as an offset in points...
            textcoords='offset points',

            # Any other text parameters we'd like
            size=14, ha='center', va='bottom')

```python
# define text string to place next to the plot
textstr= r'$E_{Tan}$ = %.2f GPa' %(tang_youngs/1000)
textstr+= "\n"
textstr+= r'$E_{Sec}$ = %.2f GPa' %(secant_youngs/1000)
textstr+= "\n"
textstr+= r'$E_{Av}$ = %.2f GPa' %(av_youngs/1000)


# add annotation next to the plot, where the top right hand side of the plot is (1, 1)
ax.annotate(textstr, xy=(1.03,0.5),
            xycoords=('axes fraction', 'figure fraction'),
            size = 14, 
            bbox=dict(boxstyle="round", alpha=0.1)) #add textbox behind text, alpha controls transparency
            
# adjust plot view so that plot takes up 70% of figure window and there is space to the right
plt.subplots_adjust(right=0.7)

# show the plot
fig
```

The final step is to graphically the Young's modulus calculations on the plot.

First up, secant Young's modulus:

```python
sec = ax.plot([0,strain_at_50],[0,df.stress[idx]],
              linestyle="dashed", color='blue',
              label=r'$E_{Sec}$') # remember the format plot([x0 x1],[y0 y1])

fig
```

Next up, we add the average Young's modulus to the plot.

```python
av = ax.plot([xmin, xmax],[df.stress[idxmin],df.stress[idxmax]],
              linestyle="dashed", color='red',
              label=r'$E_{Av}$') # remember the format plot([x0 x1],[y0 y1])

fig
```

Then we graphically add the tangential Young's modulus to the plot.

```python
tang = ax.plot([df.strain[idx-1], df.strain[idx+1]],[df.stress[idx-1],df.stress[idx+1]],
              linestyle="dashed", color='green') # remember the format plot([x0 x1],[y0 y1])

fig
```

However: this line is very small and hard to see, so we can play around with the plotting. Let's take a point on the line - the x and y coordinates are...

```python
[df.strain[idx-1], df.stress[idx-1]]
```

We also know the gradient of the line is:

```python
tang_youngs
```

The equation of a line is y=mx+c. The c intercept can therefore be calculated by:

c = y-mx

```python
c = df.stress[idx-1] - (tang_youngs*df.strain[idx-1])
print(c)
```

So if we want to plot a line, for example from 0.0028 to 0.0032, we just substitute these values for x in the equation:

```python
y1 = (tang_youngs*0.0028)+c
y2 = (tang_youngs*0.0032)+c
```

We might want to automate this if the script was going to be used for multiple datasets. But now we can plot the tangential line more clearly:

```python
tang = ax.plot([0.0028, 0.0032],[y1,y2],
              linestyle="dashed", color='green',
              label=r'$E_{Tan}$') # remember the format plot([x0 x1],[y0 y1])

fig
```
